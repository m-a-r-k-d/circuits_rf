# Amplitude modulation is multiplying

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.pi
```

```python
print(np.sin(np.pi/4))
```

```python
f_c = 1e5
```

```python
f_c
```

```python
f_mod = 1e3
```

```python
t_arr = np.arange(0,1e-5,1e-7)
```

```python
print(t_arr)
```

```python
wave_c = np.sin(2*np.pi*f_c*t_arr)
```

```python
plt.plot(t_arr, wave_c)
```

For the modulating signal, we need a t_arr that is more like 2 msec

```python
t_arr = np.arange(0,2e-3,1e-7)
```

```python
wave_c = np.sin(2*np.pi*f_c*t_arr)
```

```python
plt.plot(t_arr, wave_c)
```

That's pretty much what I'd expect for 100 kHz over 2 msec

```python
t_arr2 = np.arange(0,0.002,1e-5)
```

```python
wave_mod = np.sin(2*np.pi*f_mod*t_arr2)
```

```python
plt.plot(t_arr2, wave_mod)
```

Make a new wave_mod using t_arr

```python
wave_mod = np.sin(2*np.pi*f_mod*t_arr)
```

```python
len(wave_mod)
```

```python
mult = wave_mod*wave_c
```

```python
mult[0:10]
```

```python
plt.plot(t_arr, mult)
```
